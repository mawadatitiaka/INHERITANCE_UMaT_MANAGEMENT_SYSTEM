```
# Parent Class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")


# Child Class: Student
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    # Overriding display_info()
    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id}")

    def enroll_course(self, course):
        print(f"{self.name} has enrolled in {course}.")


# Child Class: Lecturer
class Lecturer(Person):
    def __init__(self, name, age, employee_id):
        super().__init__(name, age)
        self.employee_id = employee_id

    def teach_course(self, course):
        print(f"{self.name} is teaching {course}.")


# -----------------------------
# Create Objects
# -----------------------------
student = Student("Mawada", 21, "UMAT2025001")
lecturer = Lecturer("Dr. Cobbinah", 45, "EMP1001")

print("STUDENT INFORMATION")
student.display_info()          # Inherited and overridden method
student.enroll_course("Object-Oriented Programming")

print("LECTURER INFORMATION")
lecturer.display_info()         # Inherited method
print(f"Employee ID: {lecturer.employee_id}")
lecturer.teach_course("Object-Oriented Programming")
```
